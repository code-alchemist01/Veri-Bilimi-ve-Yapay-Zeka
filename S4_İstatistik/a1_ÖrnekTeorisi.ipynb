# Örnek Teorisi

Örnekleme teorisi, istatistikte, bir ana kütleden (popülasyon) belirli kurallara göre seçilen örnekler üzerinden çıkarımlar yapmayı inceleyen teorik çerçevedir. Ana amacı, örnekleme yoluyla elde edilen verilerden popülasyon hakkında tutarlı, yanlılıktan arınmış ve verimli tahminler yapmaktır.

### Temel Kavramlar:
1) Ana Kütle (Popülasyon): İncelenen tüm bireylerin oluşturduğu topluluk.
2) Örneklem (Sample): Ana kütleden belirli bir yöntemle seçilen alt grup.
3) Örnekleme Dağılımı: Bir istatistiğin (örneğin ortalama, varyans) farklı örneklerde nasıl dağıldığını gösteren dağılım.
4) Önyargı (Bias): Tahmin edilen değerin sistematik olarak gerçek değerden sapması.
5) Örnekleme Yöntemleri: Rastgele örnekleme, sistematik örnekleme, tabakalı örnekleme gibi teknikler.

### Önemli Teoremler ve Kavramlar:
* Merkezi Limit Teoremi (CLT): Büyük örneklem boyutlarında, örneklem ortalamasının normal dağılıma yaklaşacağını belirtir.
* Yasa Sayısı Kanunu (Law of Large Numbers, LLN): Örneklem büyüdükçe, örneklem ortalamasının popülasyon ortalamasına yaklaşacağını ifade eder.

Özetle, örnekleme teorisi, küçük bir veri kümesinden büyük bir popülasyon hakkında doğru tahminlerde bulunmayı sağlayan istatistiksel kurallar bütünüdür.

### Betimsel istatistik

Betimsel istatistik, verilerin temel özelliklerini özetlemek ve anlamak amacıyla kullanılan bir istatistik dalıdır. Bu tür istatistikler, bir veri setinin dağılımını, merkezi eğilim ölçülerini, yayılımını ve diğer temel özelliklerini tanımlar. Amaç, verinin genel yapısını anlamaktır, ancak veri ile ilgili kesin çıkarımlar yapmaz.

Betimsel istatistikte kullanılan bazı yaygın ölçütler şunlardır:

1. Merkezi Eğilim Ölçüleri:

* Aritmetik Ortalama (Ortalamayı): Verilerin toplamının, veri sayısına bölünmesiyle elde edilir.
* Medyan: Veri setini küçükten büyüğe sıraladığınızda, ortada yer alan değer.
* Mod: En fazla tekrar eden değer.

2. Dağılım Ölçüleri:

* Standart Sapma: Verilerin ortalamadan ne kadar sapma gösterdiğini belirler.
* Varyans: Verilerin ortalama etrafında ne kadar dağıldığını ölçer.
* Çeyrekler Aralığı: Veri setinin dağılımını daha iyi anlamak için ilk çeyrek (Q1) ile üçüncü çeyrek (Q3) arasındaki fark.

3. Frekans Dağılımı: Verilerin belirli bir aralığa ne sıklıkta düştüğünü gösterir.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Fiyatların dağılımını görselleştirme
plt.figure(figsize=(10,6))
sns.histplot(fiyatlar, kde=True, bins=30)
plt.title('Fiyatların Dağılımı')
plt.xlabel('Fiyatlar')
plt.ylabel('Frekans')
plt.show()

# Kutu grafiği ile dağılımı görselleştirme
plt.figure(figsize=(6,4))
sns.boxplot(x=fiyatlar)
plt.title('Fiyatların Kutu Grafiği')
plt.show()

# Fiyatların temel istatistiklerini hesaplama
mean_fiyat = fiyatlar.mean()
median_fiyat = np.median(fiyatlar)
std_fiyat = fiyatlar.std()
min_fiyat = fiyatlar.min()
max_fiyat = fiyatlar.max()

print(f"Ortalama Fiyat: {mean_fiyat}")
print(f"Medyan Fiyat: {median_fiyat}")
print(f"Standart Sapma: {std_fiyat}")
print(f"Min Fiyat: {min_fiyat}")
print(f"Max Fiyat: {max_fiyat}")
```

```python
from scipy import stats

# Shapiro-Wilk testi ile normal dağılımı kontrol etme
shapiro_test_stat, shapiro_p_value = stats.shapiro(fiyatlar)
print(f"Shapiro-Wilk Testi Sonucu: Test istatistiği = {shapiro_test_stat}, p-değeri = {shapiro_p_value}")

# Eğer p-değeri 0.05'ten küçükse, normal dağılımdan sapma vardır.
if shapiro_p_value < 0.05:
    print("Veri normal dağılımdan sapmaktadır.")
else:
    print("Veri normal dağılımdan sapmamaktadır.")
```

```python

import statsmodels.api as sm

# Talep verisi oluşturalım (örnek olarak, fiyatlara ters orantılı bir talep oluşturalım)
talep = 200 - fiyatlar + np.random.normal(0, 15, 1000)  # Talep azalan bir fonksiyon

# Regresyon analizi
X = sm.add_constant(fiyatlar)  # Fiyatları bağımsız değişken olarak ekliyoruz
model = sm.OLS(talep, X)  # OLS (En Küçük Kareler) modelini kuruyoruz
results = model.fit()

# Sonuçları yazdıralım
print(results.summary())
```

```python
import pandas as pd

# Zaman serisi verisi oluşturalım (örnek olarak 1000 gün boyunca fiyatlar)
dates = pd.date_range(start='2020-01-01', periods=1000, freq='D')
fiyatlar_series = pd.Series(fiyatlar, index=dates)

# Zaman serisi grafiğini çizelim
plt.figure(figsize=(12,6))
fiyatlar_series.plot()
plt.title('Zamanla Fiyat Değişimi')
plt.xlabel('Tarih')
plt.ylabel('Fiyat')
plt.show()

# Hareketli ortalama ile trendi gözlemleme
fiyatlar_series_rolling = fiyatlar_series.rolling(window=30).mean()  # 30 günlük hareketli ortalama
plt.figure(figsize=(12,6))
fiyatlar_series.plot(label='Fiyatlar')
fiyatlar_series_rolling.plot(label='30 Günlük Hareketli Ortalama', color='orange')
plt.title('Fiyat Değişimi ve Hareketli Ortalama')
plt.xlabel('Tarih')
plt.ylabel('Fiyat')
plt.legend()
plt.show()
```

### Olasılık Dağılımları

Olasılık Dağılımları, bir rastgele değişkenin alabileceği değerlerin ve bu değerlerin olasılıklarının dağılımını tanımlar. İki ana türü vardır:

1. Sürekli Olasılık Dağılımları: Bir değişkenin sürekli bir aralıkta alabileceği değerleri ifade eder. Örneğin:

* Normal Dağılım: Çan eğrisi şeklinde, genellikle doğal olaylarda (örneğin, boy uzunluğu) görülür.

* Üstel Dağılım: Bir olayın belirli bir sürede gerçekleşme olasılığı, örneğin bir makinenin arıza süresi.

2. Ayrık Olasılık Dağılımları: Belirli, sayılabilir değerler alır. Örneğin:

* Bernoulli Dağılımı: İki olası sonuç (başarı/başarısızlık) için kullanılır.

* Binom Dağılımı: Birden fazla Bernoulli denemesi sonucunda başarı sayısını modelleyen dağılımdır.

Her olasılık dağılımı, bir beklenen değer (ortalama), varyans ve standart sapma gibi temel özelliklere sahiptir. Olasılık yoğunluğu fonksiyonu (PDF) sürekli dağılımlar için, kümülatif dağılım fonksiyonu (CDF) ise belirli bir değere kadar olan olasılığı hesaplamak için kullanılır.

Olasılık dağılımları, istatistik, makine öğrenmesi, risk analizi gibi pek çok alanda veri analizini ve tahmin yapmayı sağlar.

### Büyük Sayılar Yasası

Büyük Sayılar Yasası, bir rastgele değişkenin olasılık dağılımının büyük sayılara yaklaştığında, olasılığın logaritmasının normal dağılımına yaklaştığını ifade eder. Deney sayısı arttıkça normal dağılıma yaklaşmış

```python
import numpy as np
rng = np.random.RandomState(123)
for i in np.arange(1,21):
    deney_sayisi = 2**i
    yazi_turalar = rng.randint(0,2,size = deney_sayisi)
    yazi_olasiliklari = np.mean(yazi_turalar)
    print("Atış Sayısı:",deney_sayisi,"---","Yazı Olasılığı: %.2f" % (yazi_olasiliklari * 100))
```

### Binom Dağılımı

Binom dağılımı, bir denemede olasılıkları aynı olan bir değişkenin olasılık dağılımıdır. Örneğin, bir olayın bir kerede başarılı olma olasılığı 0.6, diğer kerede başarısız olma olasılığı ise 0.4 olan bir denemede, olayın 3 kez başarılı olma olasılığı 0.48 dir.

### Reklam Harcaması Optimizasyonu


```python
from scipy.stats import binom
```

```python
p = 0.01
n = 100
rv = binom(n, p)
print(rv.pmf(1)) # 1 tane başarılı olma olasılığı
print(rv.pmf(5)) # 5 tane başarılı olma olasılığı
print(rv.pmf(10)) # 10 tane başarılı olma olasılığı
```

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

# Reklamın tıklanma olasılığı ve gösterim sayısı
p = 0.01  # Her bir kişinin reklama tıklama olasılığı
n = 100   # Reklamın gösterileceği toplam kişi sayısı

# Binom dağılımını oluşturuyoruz
rv = binom(n, p)

# 0'dan 100'e kadar tıklama sayılarının olasılıklarını hesapla
x = np.arange(0, n+1)
pmf_values = rv.pmf(x)

# Grafikle tıklama olasılıklarını göster
plt.figure(figsize=(10,6))
plt.bar(x, pmf_values, color='skyblue')
plt.title("Reklam Tıklama Olasılıkları (100 Kişi, p = 0.01)")
plt.xlabel("Başarı (Tıklama Sayısı)")
plt.ylabel("Olasılık")
plt.show()

# Özel tıklama sayılarının olasılıklarını yazdır
print(f"1 kişiden reklamı tıklama olasılığı: {rv.pmf(1)}")
print(f"5 kişiden reklamı tıklama olasılığı: {rv.pmf(5)}")
print(f"10 kişiden reklamı tıklama olasılığı: {rv.pmf(10)}")
```

### Poisson Dağılımı

Poisson dağılımı, belirli bir zaman aralığında bir olayın ortalama geçiren sayıyı modellemek için kullanılan dağılımdır. Örneğin, bir olayın 1 saatte ortalama 10 geçme olasılığı vardır.

### İlan girişi hata olasılıklarının Hesaplanması 


```python
from scipy.stats import poisson
```

```python
lambda_ = 0.1  # Ortalama 0.1 hata sayısı
```

```python
rv = poisson(mu=lambda_)
print(rv.pmf(k=0))  # 0 hata olma olasılığı
```

```python
print(rv.pmf(k=3))  # 3 hata olma olasılığı
```

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

# Ortalama hata sayısı
lambda_ = 0.1  

# Poisson dağılımı oluştur
rv = poisson(mu=lambda_)

# 0'dan 10'a kadar olan hata sayıları için olasılıkları hesapla
k_values = np.arange(0, 11)
pmf_values = rv.pmf(k_values)

# Grafikle Poisson dağılımını göster
plt.figure(figsize=(10,6))
plt.bar(k_values, pmf_values, color='skyblue')
plt.title("Poisson Dağılımı: Ortalama Hata Sayısı = 0.1")
plt.xlabel("Hata Sayısı")
plt.ylabel("Olasılık")
plt.xticks(k_values)  # x ekseninde tüm hata sayıları
plt.show()

# Belirli hata sayıları için olasılıkları yazdır
print(f"0 hata olma olasılığı: {rv.pmf(0)}")
print(f"3 hata olma olasılığı: {rv.pmf(3)}")
print(f"5 hata olma olasılığı: {rv.pmf(5)}")
print(f"10 hata olma olasılığı: {rv.pmf(10)}")
```

### Normal Dağılım 

Normal dağılım, bir değişkenin ortalaması ve standart sapmasının verildiği dağılımdır. Örneğin, bir insanın boy ve kilo değerlerinin bir normal dağılımına sahip olduğunu söyleyebiliriz.

### Ürün Satış olasılıklarının Hesaplanması


```python
from scipy.stats import norm
# 90'dan fazla olma olasılığı
1 - norm.cdf(90, 80, 5)
```

```python
# 70'den az olma olasılığı
norm.cdf(70, 80, 5)
```

```python
#85 ile 90 arasında olma olasılığı
norm.cdf(90, 80, 5) - norm.cdf(85, 80, 5)
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Ortalama ve standart sapma
mu = 80
sigma = 5

# Farklı x değerleri için kümülatif olasılıkları hesapla
x_values = np.linspace(60, 100, 1000)  # 60 ile 100 arasındaki değerler
cdf_values = norm.cdf(x_values, mu, sigma)

# 90'dan fazla olma olasılığı
prob_90_plus = 1 - norm.cdf(90, mu, sigma)

# 70'ten az olma olasılığı
prob_70_less = norm.cdf(70, mu, sigma)

# Grafikle normal dağılımı ve olasılıkları göster
plt.figure(figsize=(10,6))
plt.plot(x_values, cdf_values, label="CDF: Normal Dağılım")
plt.fill_between(x_values, cdf_values, where=(x_values > 90), color='red', alpha=0.5, label="90'dan fazla olma olasılığı")
plt.fill_between(x_values, cdf_values, where=(x_values < 70), color='blue', alpha=0.5, label="70'ten az olma olasılığı")
plt.title("Normal Dağılım: Ortalama = 80, Standart Sapma = 5")
plt.xlabel("X Değerleri")
plt.ylabel("Kümülatif Olasılık (CDF)")
plt.legend()
plt.grid(True)
plt.show()

# Hesaplanan olasılıkları yazdır
print(f"90'dan fazla olma olasılığı: {prob_90_plus:.4f}")
print(f"70'ten az olma olasılığı: {prob_70_less:.4f}")
```

# Hipotez Testi 

Hipotez testi, bir istatistiksel analizin temelini oluşturan, bir grup veriye dayalı olarak bir hipotezi test etme sürecidir. Temel amaç, belirli bir iddianın (hipotezin) veri tarafından desteklenip desteklenmediğini belirlemektir.

Bir hipotez testi, genellikle aşağıdaki adımlarla gerçekleştirilir:

1. Null Hipotezi (H₀) ve Alternatif Hipotezi (H₁) Kurma:
* Null Hipotezi (H₀): Test edilen temel hipotezdir ve genellikle "hiçbir fark yok" veya "etki yok" şeklinde ifade edilir. Null hipotezi, genellikle testin başlangıç varsayımıdır.
* Alternatif Hipotez (H₁): Null hipotezinin zıddı olan, test edilen yeni varsayım veya iddiadır. Alternatif hipotez, genellikle "bir fark var" veya "bir etki var" şeklinde ifade edilir.

Örneğin:

- H₀: "Yeni ilaç, eski ilaç kadar etkili değildir."
- H₁: "Yeni ilaç, eski ilaçtan daha etkili olmalıdır."
2. Test İstatistiğini Hesaplama:
Test istatistiği, verilerden elde edilen sayısal değerdir. Bu değer, genellikle örneklem ortalaması, örneklem varyansı gibi özellikler kullanılarak hesaplanır. Test istatistiği, null hipotezinin doğruluğu hakkında fikir verir.

3. P-değeri Hesaplama:
P-değeri, testin sonucunun rastgelelikten ne kadar uzak olduğunu gösterir. P-değeri, null hipotezinin doğru olduğuna dair kanıtın ne kadar zayıf veya güçlü olduğunu belirtir.

* P-değeri düşükse (genellikle 0.05'ten küçükse), null hipotezini reddetmek için yeterli kanıt vardır.
* P-değeri yüksekse (genellikle 0.05'ten büyükse), null hipotezi reddedilemez ve kabul edilir.
4. Karar Verme:
Eğer P-değeri < α (genellikle 0.05) ise, null hipotezi reddedilir ve alternatif hipotez kabul edilir.
Eğer P-değeri ≥ α ise, null hipotezi reddedilemez.
5. Sonuçları Yorumlama:
Sonuçlar, elde edilen P-değeri ve test istatistiği kullanılarak yorumlanır. Hipotez testi, sonucun anlamlı olup olmadığını belirler ve verilerin null hipotezi destekleyip desteklemediğine karar verir.

Örnek:
Bir okul, yeni bir öğretim yönteminin öğrenci başarılarını artırıp artırmadığını test etmek istiyor. Önceden yapılan bir araştırmaya göre, bu okulda öğrencilerin başarı puanlarının ortalaması 70'tir. Yeni öğretim yöntemini uygulayan bir grup öğrencinin başarı puanlarının ortalamasının 70'ten farklı olup olmadığını test etmek istiyoruz.

Adımlar:
Hipotezler:

1. H₀: Yeni öğretim yönteminin, öğrencilerin başarı puanlarını değiştirmediği, yani ortalama başarı puanı 70'tir.
2. H₁: Yeni öğretim yöntemi, öğrencilerin başarı puanlarını değiştirmiştir, yani ortalama başarı puanı 70'ten farklıdır.
3. Test İstatistiği: Öğrencilerin başarı puanlarının ortalaması hesaplanır ve bu değerin null hipotezinin (70) etrafındaki sapma hesaplanır.

P-değeri Hesaplama: Bu adımda, başarı puanlarının örneklem dağılımından elde edilen P-değeri hesaplanır.

Karar Verme: P-değeri α = 0.05'ten küçükse, H₀ reddedilir ve H₁ kabul edilir.

Hipotez Testi Türleri:
1. Z Testi: Popülasyonun standart sapması bilindiğinde ve örneklem sayısı büyük olduğunda kullanılır.
2. t Testi: Popülasyonun standart sapması bilinmediğinde ve örneklem sayısı küçük olduğunda kullanılır.
3. Ki-kare Testi: Kategorik verilerin analizinde, özellikle beklenen ve gözlemlenen frekansların karşılaştırılması için kullanılır.
4. ANOVA (Varyans Analizi): Birden fazla grup arasında ortalamaların karşılaştırılmasında kullanılır.

# Hipotez Testinde Hata Tipleri

Hipotez testlerinde, yapılan testin sonucuna göre iki tür hata meydana gelebilir. Bu hatalar, doğru karar vermek için kullanılan istatistiksel testin doğasından kaynaklanır. Hatalar **Tip I** ve **Tip II** olarak ikiye ayrılır:

## Tip I Hata (False Positive)
- **Tanım**: **Null hipotezi doğru olduğu halde reddedildiğinde** yapılan hata.
- **Açıklama**: Bu hata, aslında doğru olan bir durumu yanlış bir şekilde reddetmek anlamına gelir. Başka bir deyişle, gerçekte etkisi olmayan bir durumun, varmış gibi kabul edilmesidir.
- **Olasılık**: Tip I hata olasılığı, **α (alpha)** ile ifade edilir. Bu, kabul edilen anlamlılık seviyesidir (genellikle 0.05 olarak seçilir).
- **Örnek**: Bir ilaç şirketinin, yeni ilacın etkin olduğuna dair bir hipotezi test etmesi. Eğer ilaç aslında etkisizse ama test sonucunda etkiliymiş gibi kabul edilirse, bu Tip I hata olur.

### Tip I Hata Örneği
- Null Hipotezi (H₀): Yeni ilaç etkisizdir.
- Alternatif Hipotezi (H₁): Yeni ilaç etkilidir.
- Eğer ilaç aslında etkisizse ve biz etkilidir diyorsak, bu Tip I hata olur.

## Tip II Hata (False Negative)
- **Tanım**: **Null hipotezi yanlış olduğu halde kabul edildiğinde** yapılan hata.
- **Açıklama**: Bu hata, gerçekte etkisi olan bir durumu, etkisizmiş gibi kabul etmek anlamına gelir. Yani, aslında var olan bir etkinin gözden kaçırılmasıdır.
- **Olasılık**: Tip II hata olasılığı **β (beta)** ile ifade edilir. Tip II hata olasılığını düşürmek, testin gücünü artırmak anlamına gelir.
- **Örnek**: Bir ilaç şirketinin, yeni ilacın etkin olduğuna dair bir hipotezi test etmesi. Eğer ilaç aslında etkiliyse ama test sonucunda etkisizmiş gibi kabul edilirse, bu Tip II hata olur.

### Tip II Hata Örneği
- Null Hipotezi (H₀): Yeni ilaç etkisizdir.
- Alternatif Hipotezi (H₁): Yeni ilaç etkilidir.
- Eğer ilaç aslında etkili olduğu halde etkisizdir diye kabul edersek, bu Tip II hata olur.

## Hata Tiplerinin İlişkisi
- **α (alpha)**: Tip I hata olasılığıdır. Anlamlılık testi için belirlenen **eşik değeri**dir (genellikle 0.05).
- **β (beta)**: Tip II hata olasılığıdır. Bu hata, **testin gücü** ile ters orantılıdır.
- **Testin Gücü**: Testin, yanlış olan bir null hipotezini reddetme yeteneğidir. Yani, Tip II hatayı azaltma gücüdür.

### Özet:
- **Tip I Hata**: Gerçekten doğru olan bir hipotezi reddetmek.
- **Tip II Hata**: Gerçekten yanlış olan bir hipotezi kabul etmek.
- Bu hatalar, genellikle **testin anlamlılık seviyesi (α)** ve **gücü (1-β)** arasında bir denge kurarak minimize edilmeye çalışılır.


# Tek Örneklem T Testi

**Tek örneklem t testi**, bir örneklemin, bilinen bir popülasyon ortalamasına göre anlamlı bir fark olup olmadığını test etmek için kullanılır. Bu test, özellikle popülasyon ortalamasının bilinmediği durumlarda, küçük örneklemlerle çalışırken kullanılır.

## Hedef

Bir popülasyon ortalaması ile örneklem ortalaması arasındaki farkın istatistiksel olarak anlamlı olup olmadığını belirlemek.

## Hipotezler

- **Null Hipotezi (H0):** Örneklem ortalaması, popülasyon ortalaması ile aynıdır.  

- **Alternatif Hipotez (H1):** Örneklem ortalaması, popülasyon ortalamasından farklıdır.  

## Testin Uygulanması

Testin uygulanabilmesi için aşağıdaki şartların sağlanması gereklidir:
1. Örneklem rasgele seçilmelidir.
2. Veriler normal dağılıma yakın olmalıdır.
3. Örneklem büyüklüğü küçükse, dağılımın normal olması daha önemlidir.

### T Testinin Formülü:



### Ürün satın alma adım optimizasyonu

```python
import numpy as np
import scipy.stats as stats

# Eski fiyatla yapılan satış verileri
eski_fiyat_satislari = [200, 220, 250, 180, 230, 210, 240, 190, 225, 215]

# Yeni fiyatla yapılan satış verileri
yeni_fiyat_satislari = [210, 230, 255, 205, 235, 225, 245, 195, 230, 220]

# Eski fiyatın ortalama satış miktarı (mu_0)
mu_0 = np.mean(eski_fiyat_satislari)

# Yeni fiyatın ortalama satış miktarı (x_bar)
x_bar = np.mean(yeni_fiyat_satislari)

# Standart sapma ve örneklem büyüklüğü
s = np.std(yeni_fiyat_satislari, ddof=1)
n = len(yeni_fiyat_satislari)

# T istatistiğini hesapla
t_stat = (x_bar - mu_0) / (s / np.sqrt(n))

# Serbestlik derecesi (df) ve p-değeri hesaplanması
df = n - 1
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

# Sonuçları yazdır
print(f"T istatistiği: {t_stat:.3f}")
print(f"P-değeri: {p_value:.3f}")

# Hipotez testi sonucu
alpha = 0.05
if p_value < alpha:
    print("Null hipotez reddedilir, yeni fiyat satışlar üzerinde anlamlı bir fark yaratmıştır.")
else:
    print("Null hipotez reddedilemez, yeni fiyat satışlar üzerinde anlamlı bir fark yaratmamıştır.")
```

### Web Sitesinde geçirilen sürenin testi

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Web sitesinde geçirilen sürelerin örnek veri kümesi (dakika cinsinden)
web_sureleri = [4.8, 5.1, 5.2, 4.9, 5.5, 5.3, 4.7, 5.0, 4.6, 5.4,
                5.2, 4.8, 5.3, 5.1, 5.0, 5.2, 4.9, 5.4, 5.1, 4.7,
                5.0, 4.8, 5.5, 5.0, 5.2, 4.9, 5.3, 5.1, 4.6, 5.4]

# Adım 1: Ortalamanın hesaplanması
ortalama_sure = np.mean(web_sureleri)

# Adım 2: t-testi yapılması
t_degeri, p_degeri = stats.ttest_1samp(web_sureleri, 5)

# Grafik 1: Histogram
plt.figure(figsize=(10, 6))
plt.hist(web_sureleri, bins=8, color='skyblue', edgecolor='black')
plt.axvline(x=ortalama_sure, color='red', linestyle='dashed', linewidth=2, label=f'Ortalama = {ortalama_sure:.2f} dk')
plt.axvline(x=5, color='green', linestyle='dashed', linewidth=2, label='Beklenen Ortalama = 5 dk')
plt.title('Web Sitesinde Geçirilen Sürelerin Dağılımı')
plt.xlabel('Geçirilen Süre (Dakika)')
plt.ylabel('Frekans')
plt.legend()
plt.grid(True)
plt.show()

# Grafik 2: Kutu Grafiği (Boxplot)
plt.figure(figsize=(10, 6))
sns.boxplot(data=web_sureleri, color='lightblue')
plt.axhline(y=5, color='green', linestyle='dashed', linewidth=2, label='Beklenen Ortalama = 5 dk')
plt.title('Web Sitesinde Geçirilen Sürelerin Kutu Grafiği')
plt.xlabel('Web Sitesinde Geçirilen Süre')
plt.ylabel('Dakika')
plt.legend()
plt.grid(True)
plt.show()

# t-testi Sonuçları
print(f"t-değeri: {t_degeri:.3f}")
print(f"p-değeri: {p_degeri:.3f}")

# Sonuç Yorumlaması
alpha = 0.05  # %5 anlamlılık seviyesi
if p_degeri < alpha:
    print("Null hipotezi reddedilir: Web sitesinde geçirilen süre 5 dakikadan farklıdır.")
else:
    print("Null hipotezi reddedilemez: Web sitesinde geçirilen süre 5 dakikadır.")
```

Tek örneklem için nonparametrik testlerden biri, Wilcoxon işaretli sıralar testidir. Bu test, örneklem verilerinin merkezi eğilimde (ortalama) bir değişiklik olup olmadığını test etmek için kullanılır. Adımlar:

3. Wilcoxon İşaretli Sıralar Testi: Bu test, bir örneklemin belirli bir medyandan farklı olup olmadığını test etmek için kullanılır. Eğer verilerin dağılımı normal değilse ve sıralama yapılabiliyorsa, Wilcoxon testi yapılabilir.

# Bağımsız İki Örneklem T Testi (Independent Two-Sample T-Test)

## Tanıtım

**Bağımsız İki Örneklem T Testi** (Independent Two-Sample T-Test), iki bağımsız grup arasındaki ortalama farkını test etmek için kullanılan parametrik bir testtir. Bu test, genellikle iki farklı grubun ortalamalarının birbirinden farklı olup olmadığını anlamak için kullanılır. Örnek olarak, iki farklı tedavi yönteminin etkinliklerini karşılaştırmak, iki farklı sınıfın sınav sonuçlarını karşılaştırmak gibi senaryolarda uygulanabilir.

### Hipotezler

- **Null Hipotezi (H₀)**: İki grup arasında ortalama farkı yoktur. Yani, grupların ortalamaları birbirine eşittir.

  
- **Alternatif Hipotez (H₁)**: İki grup arasında ortalama farkı vardır. Yani, grupların ortalamaları birbirine eşit değildir.
  


Bu testin temel amacı, örneklemlerden elde edilen verilerle, iki grubun popülasyon ortalamaları arasında anlamlı bir fark olup olmadığını test etmektir.

### Bağımsız İki Örneklem T-Testi Ne Zaman Kullanılır?

- İki bağımsız grubun ortalamaları arasındaki farkı test etmek için.
- Grupların normal dağıldığı varsayımı altında.
- Her iki grup için varyansların eşit olduğu varsayımı altında (veya eşit olmayan varyanslar için farklı bir t-testi uygulanabilir).







# Nonparametrik Bağımsız İki Örneklem Testi

**Nonparametrik Bağımsız İki Örneklem Testi**, bağımsız iki grup arasındaki farkları test etmek için kullanılan bir istatistiksel testtir. Bu test, verilerin normal dağılım varsayımını karşılamadığı durumlarda kullanılır. Parametrik testler, örneğin **Bağımsız İki Örneklem T Testi** gibi, verilerin normal dağılıma uymasını gerektirirken, nonparametrik testler bu varsayımın sağlanmadığı durumlarda kullanılır. 

Nonparametrik testlerden biri olan **Mann-Whitney U Testi**, iki bağımsız örneklem arasındaki farkları incelemek için kullanılır. Bu test, grupların sıralı verileri üzerinden hesaplama yaparak, aralarındaki farkları analiz eder. 

---

# Bağımlı İki Örneklem Testi

**Bağımlı İki Örneklem Testi**, aynı denek grubunun iki farklı durumda ölçülen değerleri arasındaki farkı test etmek için kullanılır. Bu test, grupların bağımlı olduğu, yani birbirini takip eden veya eşleşen gözlemler olduğu durumlarda uygulanır. Örnek olarak, bir tedavi öncesi ve sonrası yapılan ölçümler veya aynı kişilerin iki farklı zaman dilimindeki ölçümleri gibi senaryolar kullanılabilir. 

Parametrik bir test olan **Paired T-Testi**, grupların normal dağıldığı varsayımı altında kullanılırken, normal dağılım varsayımı sağlanmadığında **Wilcoxon Signed-Rank Testi** gibi nonparametrik testler tercih edilir.

---

# Nonparametrik Bağımlı İki Örneklem Testi

**Nonparametrik Bağımlı İki Örneklem Testi**, bağımlı iki örneklem arasındaki farkları test etmek için kullanılan bir başka istatistiksel testtir. Bu test, verilerin normal dağılım göstermediği durumlarda kullanılır. Bağımlı örneklemler genellikle aynı denekler üzerinde yapılan ölçümler olduğu için, veriler arasındaki bağımlılığı göz önünde bulundurur.

**Wilcoxon Signed-Rank Testi**, nonparametrik bir test olarak, iki bağımlı örneklem arasındaki farkları sıralı veriler üzerinden analiz eder. Normal dağılım varsayımını sağlamayan ve sıralı verilerle yapılan analizlerde yaygın olarak kullanılır.

---

# İki Örneklem Oran Testi

**İki Örneklem Oran Testi**, iki bağımsız grup arasındaki oran farkını test etmek için kullanılan bir istatistiksel testtir. Genellikle iki grubun belirli bir olayın görülme oranları arasındaki farkı incelemek amacıyla kullanılır. Örneğin, bir tedavi yönteminin etkinliği veya bir ürünün müşteri memnuniyeti gibi oranlarla ilgili karşılaştırmalar yapılabilir.

Bu testin temel amacı, iki grubun başarı oranlarının birbirinden farklı olup olmadığını test etmektir. Oran testleri genellikle binom dağılımına dayalı olup, **z-test** ile yapılan hesaplamalarla sonucu elde edebiliriz.

---

# Varyans Analizi Varsayım Kontrolü

**Varyans Analizi** (ANOVA), üç veya daha fazla grubun ortalamalarının karşılaştırılması için kullanılır. Ancak, doğru sonuçlar elde edebilmek için ANOVA'nın bazı temel varsayımlarının sağlanması gerekir. Bu varsayımlar arasında **normal dağılım** ve **grup içi varyansların eşitliği** yer alır. 

**Varyansların Eşitliği Varsayımı** (homojenlik), gruplar arasındaki varyansların birbirine benzer olmasını gerektirir. Eğer bu varsayım sağlanmazsa, parametreler doğru sonuçlar vermez. Bu durumu test etmek için **Levene Testi** veya **Brown-Forsythe Testi** gibi testler kullanılır.

---

# Hipotez Testi

**Hipotez Testi**, bir hipotezin geçerliliğini test etmek amacıyla yapılan istatistiksel bir prosedürdür. Hipotez testlerinde, bir araştırma hipotezi (alternatif hipotez) ve buna karşıt olan null hipotez (H₀) belirlenir. Testin amacı, veriler üzerinden yapılan analizlerle null hipotezini reddetmek ya da reddetmemektir. 

Bir hipotez testinde genellikle **p-değeri** hesaplanır ve bu değere göre **null hipotezi** reddedilip reddedilmeyeceğine karar verilir. Eğer p-değeri belirli bir anlamlılık seviyesinden (genellikle 0.05) küçükse, null hipotezi reddedilir.

---

# Nonparametrik Hipotez Testi

**Nonparametrik Hipotez Testi**, verilerin normal dağılım göstermediği ya da parametrik testlerin varsayımlarını karşılamadığı durumlarda kullanılan testlerdir. Parametrik testler, örneğin **t-testi**, verilerin normal dağıldığı varsayımına dayanır. Ancak verilerin normal dağılmadığı durumlarda nonparametrik testler devreye girer.

Nonparametrik testler, verilerin sıralı olması durumunda da kullanılabilir. Örneğin, **Mann-Whitney U Testi**, **Wilcoxon Signed-Rank Testi** gibi testler, sıralı veriler üzerinde anlamlı farkları test etmek için kullanılır.

---

# Korelasyon Analizi

**Korelasyon Analizi**, iki veya daha fazla değişken arasındaki ilişkinin gücünü ve yönünü inceleyen bir istatistiksel analiz türüdür. Korelasyon, değişkenler arasındaki doğrusal ilişkiyi ölçer. En yaygın korelasyon testi **Pearson Korelasyon Katsayısı**'dır ve değişkenlerin birbirine ne kadar yakın bir doğrusal ilişki içinde olduklarını belirler.

Pozitif bir korelasyon, bir değişkenin arttıkça diğerinin de arttığını gösterirken, negatif bir korelasyon bir değişkenin arttıkça diğerinin azaldığını gösterir.

---

# Korelasyon Varsayım Kontrolü

Korelasyon analizini doğru bir şekilde yapabilmek için, verilerin belirli varsayımları sağlaması gerekir. Bu varsayımlar, analiz edilen değişkenlerin normal dağılım göstermesi ve doğrusal bir ilişki olması gerektiğini içerir.

**Pearson Korelasyon Testi** için bu varsayımların sağlanıp sağlanmadığı kontrol edilmelidir. Eğer veriler normal dağılım göstermiyorsa veya doğrusal ilişki yoksa, **Spearman Sıra Korelasyonu** gibi nonparametrik testler kullanılabilir.

---

# Korelasyon Katsayısı Hipotez Testi

**Korelasyon Katsayısı Hipotez Testi**, iki değişken arasındaki korelasyonun istatistiksel olarak anlamlı olup olmadığını test etmek için kullanılır. Bu testin temel amacı, örneklemde gözlemlenen korelasyon katsayısının, popülasyon korelasyon katsayısından anlamlı bir şekilde farklı olup olmadığını belirlemektir.

Bu testin yapılabilmesi için **null hipotezi** genellikle korelasyon katsayısının sıfır olduğu (yani değişkenler arasında ilişki olmadığı) şeklinde kurulur. Eğer p-değeri küçükse, null hipotezi reddedilir ve değişkenler arasında anlamlı bir ilişki olduğu kabul edilir.

---

# Nonparametrik Korelasyon Testi

**Nonparametrik Korelasyon Testi**, verilerin sıralı veya kategorik olduğu durumlarda kullanılan bir testtir. **Spearman Sıra Korelasyonu** ve **Kendall's Tau** gibi testler, iki değişken arasındaki ilişkiyi incelemek için kullanılır. Bu testler, verilerin normal dağılıma uymadığı durumlarda kullanılır.

Nonparametrik korelasyon testleri, sıralı verilerde doğrusal olmayan ilişkileri ölçer. Bu nedenle, veri normal dağılmasa bile iki değişken arasındaki ilişkiyi incelemeye olanak tanır.
