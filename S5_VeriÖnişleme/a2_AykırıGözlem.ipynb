# Aykırı Gözlem ve Aykırı Değerler

## 1. Aykırı Gözlem (Outlier) Nedir?

**Aykırı gözlem (Outlier)**, bir veri setinde diğer gözlemlerle karşılaştırıldığında anlamlı derecede farklı olan gözlemlerdir. Aykırı gözlemler, verinin genel dağılımından dışarıda yer alan ve modelin doğru öğrenmesine engel olabilen verilerdir. Aykırı gözlemler bazen hatalı veri girişlerinden kaynaklanabilirken, bazen de gerçekten nadir ancak anlamlı gözlemler olabilirler.

Aykırı gözlemler, veri analizi ve makine öğrenmesi modellerinde önemli rol oynar çünkü bu gözlemler modelin doğruluğunu ve güvenilirliğini doğrudan etkileyebilir.

## 2. Kime Göre Neye Göre Aykırı Gözlem?



Aykırı gözlem, **kime göre** ve **neye göre** aykırı olduğuna bağlı olarak farklılık gösterebilir. Bu, verinin yapısına, analiz edilen problem türüne ve kullanılan metriklere bağlı olarak değişir.

- **Kime Göre Aykırı?**: Aykırı gözlemler bir grup veri noktasıyla karşılaştırıldığında belirginleşir. Örneğin, bir şirketin çalışanlarının maaşlarını analiz ediyorsanız, diğer çalışanlarla kıyaslandığında çok yüksek veya çok düşük maaşlar aykırı gözlem olarak değerlendirilebilir. Ancak bu maaşlar şirketin genel yapısına göre aykırı olabilirken, daha geniş bir sektör verisi içinde olağan olabilir.
  
- **Neye Göre Aykırı?**: Bir gözlemin aykırı olup olmadığı, veri setindeki diğer gözlemlere göre değişir. Örneğin, bir öğrencinin sınav notu genel sınıf performansından çok farklıysa, bu not aykırı olarak değerlendirilebilir. Ancak, notun yüksek olması, sınıfın geneline kıyasla bir farklılık gösterse de, bu not sektördeki diğer okulların geneliyle karşılaştırıldığında normal olabilir.

Aykırı gözlemler, genellikle belirli bir **referans dağılımına** karşı kıyaslanır. Bu referans, istatistiksel hesaplamalar veya gözlem gruplarına dayanır.

## 3. Aykırı Değerleri Yakalamak

Aykırı gözlemleri yakalamak için kullanılan birkaç yaygın yöntem bulunmaktadır. Bu yöntemler, verinin tipine ve modelleme amacına göre değişiklik gösterebilir.

### 1. Z-Skoru (Z-Score)
Z-skoru, her bir gözlemin, veri setinin ortalamasına olan uzaklığını gösteren bir değerdir. Z-skoru belirli bir eşik değeri (genellikle 3) aştığında, bu gözlem aykırı kabul edilir.

**Formül**:


### 2. IQR (Interquartile Range)
IQR, verinin **1. çeyrek (Q1)** ve **3. çeyrek (Q3)** arasındaki farktır. Aykırı gözlemler, bu çeyreklerin dışındaki değerler olarak kabul edilir. Aykırı gözlemler, genellikle \(Q1 - 1.5 \times IQR\) ve \(Q3 + 1.5 \times IQR\) aralığının dışında yer alır.

### 3. Boxplot
Boxplot (kutu grafiği), aykırı gözlemleri görsel olarak tespit etmenin etkili bir yoludur. Aykırı gözlemler, kutunun dışındaki noktalarla gösterilir.

### 4. Scatter Plot
İki değişkenli verilerde, scatter plot (dağılım grafiği) kullanarak, gözlemler arasındaki ilişkiyi ve aykırı noktaları görsel olarak tespit edebilirsiniz.

### 5. Mahalanobis Mesafesi
Çok değişkenli verilerde aykırı gözlemleri tespit etmek için **Mahalanobis mesafesi** kullanılabilir. Bu yöntem, her bir gözlemi, tüm veri setine göre ne kadar uzak olduğunu hesaplamak için kullanılır.

## 4. Aykırı Değer Problemi Çözmek

Aykırı gözlemlerle karşılaşıldığında, birkaç farklı strateji uygulanabilir. Aykırı değerlerle başa çıkma yöntemleri, verinin doğasına ve modelleme amacına göre değişebilir.

### 1. Aykırı Gözlemleri Kaldırmak
Eğer aykırı gözlemler veri hatalarından kaynaklanıyorsa veya modelin doğruluğunu olumsuz etkiliyorsa, bu gözlemler veri setinden çıkarılabilir. Ancak, bazı aykırı gözlemler gerçekten önemli bilgiler taşıyor olabilir, bu nedenle bu işlem dikkatli yapılmalıdır.

### 2. Aykırı Gözlemleri Düzeltmek
Aykırı gözlemler bazen veri giriş hatalarından kaynaklanabilir. Yanlış ölçüm veya veri giriş hatalarını düzeltmek, bu gözlemleri normal değerlere çekmek için bir çözüm olabilir.

### 3. Aykırı Değerleri Etiketlemek
Bazı durumlarda, aykırı gözlemler veri setinde anlamlıdır ve modellemede kullanılabilir. Bu gözlemler etiketlenip özel olarak işlenebilir, böylece model bu aykırı gözlemleri dikkate alır.

### 4. Robust (Dayanıklı) Modeller Kullanmak
Bazı makine öğrenmesi algoritmaları, aykırı gözlemlerden etkilenmeyecek şekilde tasarlanmıştır. Örneğin, **Random Forest** veya **Decision Tree** gibi modeller aykırı gözlemlerle başa çıkabilir. Ayrıca, **RANSAC (Random Sample Consensus)** gibi algoritmalar, aykırı gözlemlerden etkilenmeyen lineer modeller oluşturmak için kullanılabilir.

### 5. Veriyi Dönüştürmek
Verinin dağılımı çok çarpık olduğunda (skewed), veriye **log dönüşümü** veya **kare kök dönüşümü** gibi işlemler uygulanabilir. Bu dönüşümler, aykırı gözlemlerin etkisini azaltabilir.

## 5. Çok Değişkenli Aykırı Gözlem Analizi

Çok değişkenli verilerde, bir gözlemin aykırı olup olmadığını anlamak daha karmaşık hale gelir. Bu tür durumlar için bazı yöntemler bulunmaktadır:

### 1. **Mahalanobis Mesafesi**
Çok değişkenli verilerde, bir gözlemin genel dağılımdan ne kadar uzak olduğunu anlamak için **Mahalanobis mesafesi** hesaplanabilir. Mahalanobis mesafesi, verinin ortalamadan ne kadar uzak olduğunu ölçer ve çok değişkenli verilerdeki aykırı gözlemleri tespit etmek için kullanılır.



### 2. **PCA (Principal Component Analysis)**
**PCA** kullanarak, çok değişkenli veriyi daha düşük boyutlu bir uzaya projekte edebiliriz. Bu yöntem, verinin ana bileşenlerine dayanarak, aykırı gözlemleri tespit etmek için kullanılabilir. Aykırı gözlemler, genellikle bu bileşenler üzerinde normalden çok daha fazla sapma gösterir.

### 3. **Kümeleme Yöntemleri (Clustering)**
**Kümeleme algoritmaları** (örneğin, **K-means**, **DBSCAN**) kullanılarak, çok değişkenli verilerde benzer gözlemler kümelenebilir. Bu kümelerden uzaklaşan gözlemler aykırı kabul edilebilir. Özellikle **DBSCAN** gibi yoğunluk tabanlı yöntemler, aykırı gözlemleri belirlemek için etkili olabilir.

## Sonuç

Bu noktada kullanılan yöntemler ve yaklaşımlar, veri setinin yapısına, modelleme amacına ve problem türüne bağlı olarak değişir.

# Çok değişkenli Gözlem analizi

Local Outlier Factor (LOF), veri setindeki anomalileri (aykırı değerleri) tespit etmek için kullanılan bir algoritmadır. LOF, bir noktanın komşularına göre yoğunluğunu değerlendirerek, diğer noktalardan ne kadar farklı olduğunu belirler.

Nasıl Çalışır?
* LOF, her bir veri noktasının yerel yoğunluğunu hesaplar.
* Bir noktanın yoğunluğu, belirli bir k komşusuyla karşılaştırılır.
* Eğer bir noktanın yoğunluğu komşularına göre belirgin şekilde düşükse, bu nokta aykırı (outlier) olarak işaretlenir.

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds = diamonds.select_dtypes(include = ['float64', 'int64'])
df = diamonds.copy()
df = df.dropna()
df.head()
```

```python
import numpy as np 
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1) # n_neighbors = komşuluk sayısı, contamination = aykırı değerlerin oranı
clf.fit_predict(df) # Aykırı değerleri bulduk
```

```python
df_scores = clf.negative_outlier_factor_ # Aykırı değerlerin skorları
df_scores[0:10]
```

```python
np.sort(df_scores)[0:20] # Aykırı değerlerin skorlarını sıraladık
```

```python
esik_deger = np.sort(df_scores)[13] # Eşik değeri belirledik
aykiri_tf = df_scores > esik_deger # Eşik değerinden büyük olanları aykırı değer olarak belirledik
aykiri_tf
```

```python
yeni_df = df[df_scores > esik_deger] # Aykırı değerleri çıkardık
yeni_df # Aykırı değerler çıkarılmış hali
```

```python
df[df_scores < esik_deger] # Aykırı olan değerler
```

```python
df[df_scores == esik_deger] # Eşik değer
```

```python
baski_deger = df[df_scores == esik_deger] # Eşik değeri baskı değeri olarak belirledik
aykirilar = df[~aykiri_tf] # Aykırı değerleri belirledik
res = aykirilar.to_records(index = False) 
res[:] = baski_deger.to_records(index = False) # Aykırı değerlerin yerine baskı değeri atadık
res
```

```python
import pandas as pd
df[~aykiri_tf] = pd.DataFrame(res, index = df[~aykiri_tf].index) # Aykırı değerlerin yerine baskı değeri atadık
df[~aykiri_tf]
```
