# Aykr Gzlem ve Aykr Deerler

## 1. Aykr Gzlem (Outlier) Nedir?

**Aykr gzlem (Outlier)**, bir veri setinde dier gzlemlerle karlatrldnda anlaml derecede farkl olan gzlemlerdir. Aykr gzlemler, verinin genel dalmndan darda yer alan ve modelin doru renmesine engel olabilen verilerdir. Aykr gzlemler bazen hatal veri girilerinden kaynaklanabilirken, bazen de gerekten nadir ancak anlaml gzlemler olabilirler.

Aykr gzlemler, veri analizi ve makine renmesi modellerinde nemli rol oynar nk bu gzlemler modelin doruluunu ve gvenilirliini dorudan etkileyebilir.

## 2. Kime Gre Neye Gre Aykr Gzlem?



Aykr gzlem, **kime gre** ve **neye gre** aykr olduuna bal olarak farkllk gsterebilir. Bu, verinin yapsna, analiz edilen problem trne ve kullanlan metriklere bal olarak deiir.

- **Kime Gre Aykr?**: Aykr gzlemler bir grup veri noktasyla karlatrldnda belirginleir. rnein, bir irketin alanlarnn maalarn analiz ediyorsanz, dier alanlarla kyaslandnda ok yksek veya ok dk maalar aykr gzlem olarak deerlendirilebilir. Ancak bu maalar irketin genel yapsna gre aykr olabilirken, daha geni bir sektr verisi iinde olaan olabilir.
  
- **Neye Gre Aykr?**: Bir gzlemin aykr olup olmad, veri setindeki dier gzlemlere gre deiir. rnein, bir rencinin snav notu genel snf performansndan ok farklysa, bu not aykr olarak deerlendirilebilir. Ancak, notun yksek olmas, snfn geneline kyasla bir farkllk gsterse de, bu not sektrdeki dier okullarn geneliyle karlatrldnda normal olabilir.

Aykr gzlemler, genellikle belirli bir **referans dalmna** kar kyaslanr. Bu referans, istatistiksel hesaplamalar veya gzlem gruplarna dayanr.

## 3. Aykr Deerleri Yakalamak

Aykr gzlemleri yakalamak iin kullanlan birka yaygn yntem bulunmaktadr. Bu yntemler, verinin tipine ve modelleme amacna gre deiiklik gsterebilir.

### 1. Z-Skoru (Z-Score)
Z-skoru, her bir gzlemin, veri setinin ortalamasna olan uzakln gsteren bir deerdir. Z-skoru belirli bir eik deeri (genellikle 3) atnda, bu gzlem aykr kabul edilir.

**Forml**:


### 2. IQR (Interquartile Range)
IQR, verinin **1. eyrek (Q1)** ve **3. eyrek (Q3)** arasndaki farktr. Aykr gzlemler, bu eyreklerin dndaki deerler olarak kabul edilir. Aykr gzlemler, genellikle \(Q1 - 1.5 \times IQR\) ve \(Q3 + 1.5 \times IQR\) aralnn dnda yer alr.

### 3. Boxplot
Boxplot (kutu grafii), aykr gzlemleri grsel olarak tespit etmenin etkili bir yoludur. Aykr gzlemler, kutunun dndaki noktalarla gsterilir.

### 4. Scatter Plot
ki deikenli verilerde, scatter plot (dalm grafii) kullanarak, gzlemler arasndaki ilikiyi ve aykr noktalar grsel olarak tespit edebilirsiniz.

### 5. Mahalanobis Mesafesi
ok deikenli verilerde aykr gzlemleri tespit etmek iin **Mahalanobis mesafesi** kullanlabilir. Bu yntem, her bir gzlemi, tm veri setine gre ne kadar uzak olduunu hesaplamak iin kullanlr.

## 4. Aykr Deer Problemi zmek

Aykr gzlemlerle karlaldnda, birka farkl strateji uygulanabilir. Aykr deerlerle baa kma yntemleri, verinin doasna ve modelleme amacna gre deiebilir.

### 1. Aykr Gzlemleri Kaldrmak
Eer aykr gzlemler veri hatalarndan kaynaklanyorsa veya modelin doruluunu olumsuz etkiliyorsa, bu gzlemler veri setinden karlabilir. Ancak, baz aykr gzlemler gerekten nemli bilgiler tayor olabilir, bu nedenle bu ilem dikkatli yaplmaldr.

### 2. Aykr Gzlemleri Dzeltmek
Aykr gzlemler bazen veri giri hatalarndan kaynaklanabilir. Yanl lm veya veri giri hatalarn dzeltmek, bu gzlemleri normal deerlere ekmek iin bir zm olabilir.

### 3. Aykr Deerleri Etiketlemek
Baz durumlarda, aykr gzlemler veri setinde anlamldr ve modellemede kullanlabilir. Bu gzlemler etiketlenip zel olarak ilenebilir, bylece model bu aykr gzlemleri dikkate alr.

### 4. Robust (Dayankl) Modeller Kullanmak
Baz makine renmesi algoritmalar, aykr gzlemlerden etkilenmeyecek ekilde tasarlanmtr. rnein, **Random Forest** veya **Decision Tree** gibi modeller aykr gzlemlerle baa kabilir. Ayrca, **RANSAC (Random Sample Consensus)** gibi algoritmalar, aykr gzlemlerden etkilenmeyen lineer modeller oluturmak iin kullanlabilir.

### 5. Veriyi Dntrmek
Verinin dalm ok arpk olduunda (skewed), veriye **log dnm** veya **kare kk dnm** gibi ilemler uygulanabilir. Bu dnmler, aykr gzlemlerin etkisini azaltabilir.

## 5. ok Deikenli Aykr Gzlem Analizi

ok deikenli verilerde, bir gzlemin aykr olup olmadn anlamak daha karmak hale gelir. Bu tr durumlar iin baz yntemler bulunmaktadr:

### 1. **Mahalanobis Mesafesi**
ok deikenli verilerde, bir gzlemin genel dalmdan ne kadar uzak olduunu anlamak iin **Mahalanobis mesafesi** hesaplanabilir. Mahalanobis mesafesi, verinin ortalamadan ne kadar uzak olduunu ler ve ok deikenli verilerdeki aykr gzlemleri tespit etmek iin kullanlr.



### 2. **PCA (Principal Component Analysis)**
**PCA** kullanarak, ok deikenli veriyi daha dk boyutlu bir uzaya projekte edebiliriz. Bu yntem, verinin ana bileenlerine dayanarak, aykr gzlemleri tespit etmek iin kullanlabilir. Aykr gzlemler, genellikle bu bileenler zerinde normalden ok daha fazla sapma gsterir.

### 3. **Kmeleme Yntemleri (Clustering)**
**Kmeleme algoritmalar** (rnein, **K-means**, **DBSCAN**) kullanlarak, ok deikenli verilerde benzer gzlemler kmelenebilir. Bu kmelerden uzaklaan gzlemler aykr kabul edilebilir. zellikle **DBSCAN** gibi younluk tabanl yntemler, aykr gzlemleri belirlemek iin etkili olabilir.

# ok deikenli Gzlem analizi

Local Outlier Factor (LOF), veri setindeki anomalileri (aykr deerleri) tespit etmek iin kullanlan bir algoritmadr. LOF, bir noktann komularna gre younluunu deerlendirerek, dier noktalardan ne kadar farkl olduunu belirler.

Nasl alr?
* LOF, her bir veri noktasnn yerel younluunu hesaplar.
* Bir noktann younluu, belirli bir k komusuyla karlatrlr.
* Eer bir noktann younluu komularna gre belirgin ekilde dkse, bu nokta aykr (outlier) olarak iaretlenir.

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds = diamonds.select_dtypes(include = ['float64', 'int64'])
df = diamonds.copy()
df = df.dropna()
df.head()
```

```python
import numpy as np 
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1) # n_neighbors = komuluk says, contamination = aykr deerlerin oran
clf.fit_predict(df) # Aykr deerleri bulduk
```

```python
df_scores = clf.negative_outlier_factor_ # Aykr deerlerin skorlar
df_scores[0:10]
```

```python
np.sort(df_scores)[0:20] # Aykr deerlerin skorlarn sraladk
```

```python
esik_deger = np.sort(df_scores)[13] # Eik deeri belirledik
aykiri_tf = df_scores > esik_deger # Eik deerinden byk olanlar aykr deer olarak belirledik
aykiri_tf
```

```python
yeni_df = df[df_scores > esik_deger] # Aykr deerleri kardk
yeni_df # Aykr deerler karlm hali
```

```python
df[df_scores < esik_deger] # Aykr olan deerler
```

```python
df[df_scores == esik_deger] # Eik deer
```

```python
baski_deger = df[df_scores == esik_deger] # Eik deeri bask deeri olarak belirledik
aykirilar = df[~aykiri_tf] # Aykr deerleri belirledik
res = aykirilar.to_records(index = False) 
res[:] = baski_deger.to_records(index = False) # Aykr deerlerin yerine bask deeri atadk
res
```

```python
import pandas as pd
df[~aykiri_tf] = pd.DataFrame(res, index = df[~aykiri_tf].index) # Aykr deerlerin yerine bask deeri atadk
df[~aykiri_tf]
```
