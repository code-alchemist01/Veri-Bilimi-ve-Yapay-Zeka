```python
# Birleştirme (Join) işlemleri

m = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(m, columns = ["s1","s2","s3"])
df1
```

```python
df2
```

```python
pd.concat([df1,df2], ignore_index=False) # index'i sildiği için index'i tekrar sıfırlar
```

```python
pd.concat([df1,df2], ignore_index=True) # df1 ve df2'yi birleştirir. index devam eder

```

```python
df1.columns
```

```python
df2.columns = ["s1","s2","sayi3"]
df2
```

```python
pd.concat([df1,df2])
```

# Gruplama İşlemleri

```python
df = pd.DataFrame({"grup": ["A","B","C","A","B","C"], "deger": [10,20,30,40,50,60]}, columns = ["grup","deger"])
df


```

```python
df.groupby("grup") # df'yi grup sütununa göre gruplar
```

```python
df.groupby("grup").mean() # df'yi grup sütununa göre gruplar ve her grup için deger sütununun ortalamasını alır
```

```python
df.groupby("grup").sum() # df'yi grup sütununa göre gruplar ve her grup için deger sütununun toplamını alır
```

```python
df.groupby("grup").agg(["mean","sum"]) # df'yi grup sütununa göre gruplar ve her grup için deger sütununun ortalamasını ve toplamını alır
```

# Aggregation İşlemleri


```python
df = pd.DataFrame({"grup": ["A","B","C","A","B","C"], "d1": [10,20,30,40,50,60], "d2": [24,23,54,32,87,91]}, columns = ["grup","d1","d2"])
df
```

```python
df.groupby("grup").mean()
```

```python
df.groupby("grup").aggregate(["min","median","max"]) #np.median gelecek sürümde hata vereceği için median kullanılıyor
```

```python
df.groupby("grup").aggregate({"d1":["min"], "d2":["max"]})
```

# Filter İşlemleri


```python
df = pd.DataFrame({"grup": ["A","B","C","A","B","C"], "d1": [10,23,33,22,11,99], "d2": [100,253,333,262,111,969]}, columns = ["grup","d1","d2"])
df
```

```python
def filter_func(x): # d1 sütununun standart sapması 9'dan büyük olan grupları seçer
    return x["d1"].std() > 9
```

```python
df.groupby("grup").std() # df'yi grup sütununa göre gruplar ve her grup için d1 sütununun standart sapmasını alır
```

```python
df.groupby("grup").filter(filter_func) # df'yi grup sütununa göre gruplar ve filter_func fonksiyonuna göre filtreler
```

# Transform İşlemleri


```python
df = pd.DataFrame({"grup": ["A","B","C","A","B","C"], "d1": [10,23,33,22,11,99], "d2": [100,253,333,262,111,969]}, columns = ["grup","d1","d2"])
df
```

```python
df["d1"]*11 # df'deki d1 sütununun her elemanını 11 ile çarpar
```

```python
df
```

```python
df_a = df.iloc[:,1:3]
```

```python
df_a.transform(lambda x: (x - x.mean()) / x.std())  # df_a daki her elemanın ortalamasını alır ve her elemandan çıkarır. sonra standart sapmasına böler
```

```python
df.groupby("grup").apply(np.mean) # df'yi grup sütununa göre gruplar ve her grup için her elemanın ortalamasını alır
```

# Pivot Tabloları

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.groupby("sex")["survived"].mean()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack() # pivot tablo oluşturur. sex ve class sütunlarına göre survived sütununu gruplar ve mean ile ortalamasını alır. unstack ile sütunları döndürür
```

```python
# pivot ile tablo

titanic.pivot_table("survived", index = "sex", columns = "class") # pivot tablo oluşturur. sex ve class sütunlarına göre survived sütununu gruplar ve mean ile ortalamasını alır. unstack ile sütunları döndürür
```

```python
titanic.age.head() # age sütununu görüntüler

```

```python
age = pd.cut(titanic.age, [0,18,90]) # age sütununu 0-18 ve 18-90 aralıklarına ayırır
age.head(10)
```

```python
titanic.pivot_table("survived", ["sex", age], "class") # pivot tablo oluşturur. sex ve age sütunlarına göre class sütununu gruplar ve survived sütununu mean ile ortalamasını alır
```

```python
titanic.pivot_table("survived", ["sex", "age"], "class") # pivot tablo oluşturur. sex ve age sütunlarına göre class sütununu gruplar ve survived sütununu mean ile ortalamasını alır
```
