# Gözlem ve Değişken Seçimleri


```python
import pandas as pd
import numpy as np
m = np.random.randint(1,30, size = (10,3))
df = pd.DataFrame(m, columns = ["s1","s2","s3"])
df
```

```python
# loc: tanımlandığı gibi seçer

df.loc[0:3, ["s1","s2"]] # 0. ve 3. satır ve s1 ve s2 sütunu
```

```python
# iloc : alıştığımız gibi index'e göre seçer


df.iloc[0:3, 0:2] # 0. ve 3. satır ve 0. ve 2. sütun. locdan farkı index'e göre seçer
```

```python
df.iloc[:3,:2]
```

```python
# Birleştirme (Join) işlemleri

m = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(m, columns = ["s1","s2","s3"])
df1
```

```python
df2 = df1 + 192 # df1'in her elemanına 192 eklenir
```

```python
df2
```

```python
pd.concat([df1,df2]) # df1 ve df2'yi birleştirir
```

```python
pd.concat([df1,df2], ignore_index=False) # index'i sildiği için index'i tekrar sıfırlar
```

```python
pd.concat([df1,df2], ignore_index=True) # df1 ve df2'yi birleştirir. index devam eder

```

```python
df1.columns
```

```python
df2.columns = ["s1","s2","sayi3"]
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = "inner") # sadece ortak olan sütunları birleştirir
```

```python
pd.concat([df1, df2.reindex(columns=df1.columns)]) # df2'nin sütunlarını df1'in sütunlarına göre birleştirir
```

```python
# İleri Düzey Birleştirme İşlemleri

# Birebir Birleştirme

df1 = pd.DataFrame({"calisanlar": ["ali","veli","ayse","fatma"], "grup": ["Muhasebe","Muhendislik","Muhendislik","İK"]})
df1
```

```python
df2 = pd.DataFrame({"calisanlar": ["ali","veli","ayse","fatma"], "ilk_giris": [2010,2009,2014,2019]})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, on = "calisanlar") # calisanlar sütununa göre birleştirir
```

```python
# coktan teke birleştirme

df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.DataFrame({"grup": ["Muhasebe","Muhendislik","İK"], "mudur": ["Caner","Suleyman","Ayse"]})
df4
```

```python
pd.merge(df3,df4)
```

```python
# Çoktan çok birleştirme

df5 = pd.DataFrame({"grup": ["Muhasebe","Muhasebe","Muhendislik","Muhendislik","İK","İK"], "yetenekler": ["matematik","wxcel","Python","R","excel","yonetim"]})
df5


```

```python

pd.merge(df3,df5, on = "grup", how = "left") # df3'ün grup sütununa göre df5'i birleştirir
```

```python
pd.merge(df1,df5)
```
