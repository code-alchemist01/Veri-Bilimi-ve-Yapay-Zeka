```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.displot(df.price, kde = False)
# bu kod price deÄŸiÅŸkeninin daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶sterir. kde = False ile Ã§izgiyi kaldÄ±rdÄ±k.
```

```python
help(sns.displot)
```

```python
sns.displot(df.price, bins = 1000, kde = False)
```

```python
sns.displot(df.price, bins = 10, kde = False)
```

```python
sns.displot(df.price, kde = True)
# bu kodun aÃ§Ä±klamasÄ± : kde = True ile Ã§izgiyi ekledik.
```

```python
df["price"].describe()
```

```python
sns.displot(df.price, kind="kde")
# YoÄŸunluk grafiÄŸi Ã§izdik.
```

```python
sns.kdeplot(df.price, shade = True)
# YoÄŸunluk grafiÄŸi Ã§izdik. shade = True ile renklendirdik.
```
