```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["timepoint"].describe().T
```

```python
df["signal"].describe().T
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe().T
```

```python
sns.lineplot(x = "timepoint", y = "signal", data = df)
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue="event",data = df)
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue="event",style="event",data = df)
# style parametresi ile event deÄŸiÅŸkenine gÃ¶re Ã§izgi stili belirlenir.
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue="region",
             style="event",
             data = df)

# markers parametresi ile noktalarÄ± belirleriz. 
# dashes parametresi ile Ã§izgileri belirleriz.
```
