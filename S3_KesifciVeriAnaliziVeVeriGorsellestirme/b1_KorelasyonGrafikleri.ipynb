# Korelasyon Grafikleri 

#### Scatterplot
* Scatterplot, iki sürekli değişken arasındaki ilişkiyi görselleştirmek için kullanılır.
Her bir nokta, iki değişkenin kesişim noktasını temsil eder ve bu sayede veriler arasındaki korelasyon veya trend gözlemlenebilir.


##### Veri Seti Hikayesi

* total_bill: Yemeğin toplam fiyatı (bahşişler ve vergiler dahil)
* tip: Bahşiş
* sex: Ücreti ödeyen kişinin cinsiyeti(0=male, 1=female)
* smoker: Grupta sigara içen var mı? (0=No, 1=Yes)
* day: Gün (3=Thursday, 4=Friday, 5=Saturday, 6=Sunday)
* time: Ne zaman? (0=Day, 1=Night)
* size: grupta kaç kişi var?

```python
import seaborn as sns
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df)
```

# Çaprazlamalar

```python
sns.scatterplot(x = "total_bill", y = "tip", hue="time" ,data = df)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue="time", style="time", data = df)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue="day", style="time", data = df)
```

```python
sns.scatterplot(x = "total_bill", y = "tip",hue="size", size="size", data = df)
# size parametresi ile boyutlandırma yapılabilir. Bu durumda size parametresi ile belirtilen değişkenin değerleri büyüklüğü belirler.
```

# Doğrusal İlişkinin Gösterilmesi

Veri biliminde iki değişken arasındaki doğrusal ilişkiyi göstermek için regression plot (Regplot) ve linear model plot (Lmplot) kullanılır.

🔹 1. Regplot (Regresyon Grafiği):
Regplot, veri noktalarının dağılımını ve üzerine çizilen doğrusal regresyon çizgisini gösterir.
Eğer iki değişken arasında lineer (doğrusal) bir ilişki varsa, bu çizgi verinin trendini gösterir.

🔹 2. Lmplot (Lineer Model Grafiği):
Regplot’un gelişmiş versiyonu olan lmplot(), kategorik değişkenlerle birlikte regresyon analizini görselleştirmek için kullanılır.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill", y="tip", data=df)
# lmplot fonksiyonu ile regresyon çizgisi ekleyebiliriz.
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",data=df)
# hue parametresi ile smoker değişkenine göre renklendirme yapabiliriz.
# artış şiddeti ve yönü hakkında bilgi verir.
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time",data=df)
# col parametresi ile zaman değişkenine göre ayrı ayrı grafikler oluşturabiliriz.
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time",row="sex",data=df)
# row ve col parametreleri ile cinsiyet ve zaman değişkenlerine göre 4 farklı grafik oluşturabiliriz.
```
