# Violin GrafiÄŸi

Violin plot, boxplot ile yoÄŸunluk grafiÄŸinin birleÅŸimidir. Bir deÄŸiÅŸkenin daÄŸÄ±lÄ±mÄ±nÄ± ve olasÄ±lÄ±k yoÄŸunluÄŸunu (density) gÃ¶sterir.

ðŸ“Œ Violin Plot'un Temel BileÅŸenleri
* Boxplot ile Benzer Ã‡eyrekler (Q1, Q2, Q3) â†’ MedyanÄ± ve Ã§eyrekleri iÃ§erir.

* Simetrik veya Asimetrik Olabilir â†’ Normal daÄŸÄ±lÄ±mda simetriktir, farklÄ± yoÄŸunluklarda asimetrik olabilir.

```python
# Violin GrafiÄŸi
# Violin GrafiÄŸi, kutu grafiÄŸine benzer bir gÃ¶rselleÅŸtirme tekniÄŸidir. DaÄŸÄ±lÄ±mÄ±n yoÄŸunluÄŸunu gÃ¶sterir. Kutu grafiÄŸe gÃ¶re daha fazla bilgi saÄŸlar.
df.head()
```
