## Makine renmesi Trleri

1. **Denetimli renme (Supervised Learning)**:
   - Bu trde, eitim verileri belirli etiketler ile birlikte salanr. Model, giri verisi ile doru sonular eletirerek renir.
   - **rnekler**: Lojistik regresyon, karar aalar, destek vektr makineleri (SVM), yapay sinir alar.
   - **Uygulamalar**: Snflandrma (rnein, e-posta spam filtresi), regresyon (rnein, konut fiyat tahmini).

2. **Denetimsiz renme (Unsupervised Learning)**:
   - Bu trde, veriler etiketlenmemitir ve model, veri kmesindeki gizli yaplar veya kalplar renmeye alr.
   - **rnekler**: Kmeleme algoritmalar (K-means, DBSCAN), boyut indirgeme (PCA, t-SNE).
   - **Uygulamalar**: Mteri segmentasyonu, anomali tespiti, zellik karm.

3. **Pekitirmeli renme (Reinforcement Learning)**:
   - Bu trde, bir ajan evreyle etkileime girer ve dller alarak renir. Ajan, evreden ald geri bildirime gre stratejiler gelitirir.
   - **rnekler**: Q-learning, Deep Q-Networks (DQN).
   - **Uygulamalar**: Oyun oynama, robotik, otonom aralar.

Makine renmesinin baars, doru veri ve uygun model seimlerine dayanr. renme sreci, verinin doasna, kullanlan algoritmaya ve hedeflenen sonuca bal olarak farkllk gsterir.

#  Terminoloji

Makine renmesi terminolojisi, doru model tasarm ve baarl bir ekilde eitim yapabilmek iin olduka nemlidir. Bu terminoloji, algoritmalarn nasl altn anlamamza ve modellerin ktlarndan doru sonular karabilmemize olanak tanr.

### Temel Terimler:

1. **Eitim Verisi (Training Data)**:
   - Modelin renmesi iin kullanlan veridir. Bu veri, modelin parametrelerini belirler. Model bu veri zerinde eitim alr.
   
2. **Test Verisi (Test Data)**:
   - Eitimden sonra modelin performansn deerlendirmek iin kullanlan veri kmesidir. Bu veri modeli "gerek dnya" koullarnda test eder.

3. **Validation Data**:
   - Modelin eitim srecinde, hiperparametre ayarlarn yapabilmek iin kullanlan veri kmesidir. Bu veri, eitim srasnda modelin doruluunu izlemek iin kullanlr.

4. **zellik (Feature)**:
   - Modelin tahmin yapabilmesi iin kullanlan bamsz deikenlerdir. Bir zellik, verinin bir boyutunu veya niteliini temsil eder (rnein, bir evin fiyatn tahmin ediyorsak, "evin bykl" bir zellik olabilir).

5. **Etiket (Label)**:
   - Denetimli renmede modelin renmeye alt hedef deiken veya ktr. rnein, ev fiyatlar bir etiket olabilir.

6. **Ar Uyum (Overfitting)**:
   - Modelin, eitim verisindeki grlty veya rastlantsal desenleri renmesi, bylece test verisinde dk performans sergilemesi durumudur. Model, eitim verisine ok iyi uyum saladnda, genellikle genellenebilirlik kayb yaanr.

7. **Yetersiz Uyum (Underfitting)**:
   - Modelin, eitim verisinin temel yapsn renememesi durumudur. Modelin basitlii, veriyi yeterince temsil etmemesine ve dk performans sergilemesine yol aar.

8. **Hiperparametre (Hyperparameter)**:
   - Modelin eitim srecini yneten ve dardan belirlenen parametrelerdir. rnein, renme oran veya karar aac derinlii gibi ayarlar hiperparametrelerdir.

9. **zellik Mhendislii (Feature Engineering)**:
   - Ham verilerden, modelin daha iyi renmesini salayacak yeni zellikler tretme srecidir.

#  Model Dorulama Yntemleri

Model dorulama, makine renmesi modellerinin genellenebilirliini deerlendirme srecidir. Bu, modelin yalnzca eitim verisine deil, yeni ve bilinmeyen verilere kar da ne kadar iyi performans gsterdiini test etmeye ynelik bir admdr.

### Yaygn Dorulama Yntemleri:

1. **Holdout Yntemi**:
   - Veri kmesi, eitim ve test veri kmesine blnr. Model eitim veri kmesinde eitilir, test veri kmesinde ise dorulanr. 
   - **Dezavantaj**: Eitim ve test verisi arasndaki blnme sonucu dorulama sreci zaman zaman yanltc olabilir.

2. **K-Fold Cross-Validation**:
   - Veri kmesi K eit paraya blnr. Model, K-1 para ile eitilir ve kalan bir para ile dorulanr. Bu ilem K kez tekrarlanr ve her seferinde dorulama sonucu kaydedilir.
   - **Avantaj**: Kmelenmi veri setlerinde dorulama iin daha gvenilir sonular verir.

3. **Leave-One-Out Cross-Validation (LOOCV)**:
   - Bu yntem, kk veri setleri iin uygundur. Veri kmesindeki her rnek, srasyla test verisi olarak kullanlr ve geri kalan rneklerle eitim yaplr.
   - **Avantaj**: Veriyi en iyi ekilde kullanma frsat tanr, ancak hesaplama asndan pahaldr.

4. **Stratified K-Fold Cross-Validation**:
   - zellikle dengesiz snflandrma problemleri iin kullanlr. Veri setinin her snfn, her katmanda eit ekilde temsil eder.
   - **Avantaj**: Modelin tm snflarda tutarl performans sergilemesini salar.

#  Yanllk - Varyans Dei Tokuu

Yanllk ve varyans, modelin doruluu zerinde nemli etkiler yapar. Bu iki faktr arasnda bir denge bulmak nemlidir.

### Yanllk (Bias) ve Varyans (Variance) likisi:

- **Yanllk (Bias)**: Modelin, veriyi ne kadar iyi renemediinin bir lsdr. Yksek yanllk, modelin "underfitting" yapmasna neden olur.
  
- **Varyans**: Modelin eitim verisine ne kadar duyarl olduunun lsdr. Yksek varyans, modelin "overfitting" yapmasna yol aar.

### Yanllk - Varyans Dei Tokuu:

- **Yksek Yanllk, Dk Varyans**: Model ok basit ve genelletirilebilir, ancak eitim verisinde dk performans gsterir. (Underfitting)
  
- **Dk Yanllk, Yksek Varyans**: Model eitim verisine ok iyi uyum salar ancak test verisi zerinde kt performans gsterir. (Overfitting)

Optimal model, dk yanllk ve dk varyans seviyesine ulamay hedefler, bu da hem doru hem de genelletirilebilir bir modelin oluturulmasn salar.

# Parametre, Hiperparametre ve Model Optimizasyonu

### Parametreler:

- Modelin renmesi srasnda belirledii deerlerdir. rnein, dorusal regresyon modelindeki **arlklar** ve **bias terimleri** parametrelerdir.

### Hiperparametreler:

### Model Optimizasyonu:

Model optimizasyonu, hiperparametrelerin doru ekilde ayarlanmasn ifade eder. Grid Search, Random Search ve Bayesian Optimization gibi yntemler ile en iyi hiperparametreler bulunabilir.

#### Grid Search:
