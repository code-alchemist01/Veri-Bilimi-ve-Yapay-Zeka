#  Makine Öğrenmesine Giriş

Makine öğrenmesi (ML), bilgisayarların açıkça programlanmadan, verilerden öğrenmesine ve bu bilgiyi kullanarak kararlar almasına olanak tanır. Yapay zekanın bir dalı olan makine öğrenmesi, algoritmalar aracılığıyla veri analizi yapar ve bu analizleri kullanarak tahminlerde bulunur. Makine öğrenmesi, özellikle büyük veri ve karmaşık sistemlerin analizinde oldukça güçlüdür. Günümüzde makine öğrenmesi; finansal analizlerden, sağlık sektörü uygulamalarına, doğal dil işlemeye, görsel tanımaya kadar birçok alanda kullanılmaktadır.

## Makine Öğrenmesi Türleri

1. **Denetimli Öğrenme (Supervised Learning)**:
   - Bu türde, eğitim verileri belirli etiketler ile birlikte sağlanır. Model, giriş verisi ile doğru sonuçları eşleştirerek öğrenir.
   - **Örnekler**: Lojistik regresyon, karar ağaçları, destek vektör makineleri (SVM), yapay sinir ağları.
   - **Uygulamalar**: Sınıflandırma (örneğin, e-posta spam filtresi), regresyon (örneğin, konut fiyat tahmini).

2. **Denetimsiz Öğrenme (Unsupervised Learning)**:
   - Bu türde, veriler etiketlenmemiştir ve model, veri kümesindeki gizli yapıları veya kalıpları öğrenmeye çalışır.
   - **Örnekler**: Kümeleme algoritmaları (K-means, DBSCAN), boyut indirgeme (PCA, t-SNE).
   - **Uygulamalar**: Müşteri segmentasyonu, anomali tespiti, özellik çıkarımı.

3. **Pekiştirmeli Öğrenme (Reinforcement Learning)**:
   - Bu türde, bir ajan çevreyle etkileşime girer ve ödüller alarak öğrenir. Ajan, çevreden aldığı geri bildirime göre stratejiler geliştirir.
   - **Örnekler**: Q-learning, Deep Q-Networks (DQN).
   - **Uygulamalar**: Oyun oynama, robotik, otonom araçlar.

Makine öğrenmesinin başarısı, doğru veri ve uygun model seçimlerine dayanır. Öğrenme süreci, verinin doğasına, kullanılan algoritmaya ve hedeflenen sonuca bağlı olarak farklılık gösterir.

#  Terminoloji

Makine öğrenmesi terminolojisi, doğru model tasarımı ve başarılı bir şekilde eğitim yapabilmek için oldukça önemlidir. Bu terminoloji, algoritmaların nasıl çalıştığını anlamamıza ve modellerin çıktılarından doğru sonuçlar çıkarabilmemize olanak tanır.

### Temel Terimler:

1. **Eğitim Verisi (Training Data)**:
   - Modelin öğrenmesi için kullanılan veridir. Bu veri, modelin parametrelerini belirler. Model bu veri üzerinde eğitim alır.
   
2. **Test Verisi (Test Data)**:
   - Eğitimden sonra modelin performansını değerlendirmek için kullanılan veri kümesidir. Bu veri modeli "gerçek dünya" koşullarında test eder.

3. **Validation Data**:
   - Modelin eğitim sürecinde, hiperparametre ayarlarını yapabilmek için kullanılan veri kümesidir. Bu veri, eğitim sırasında modelin doğruluğunu izlemek için kullanılır.

4. **Özellik (Feature)**:
   - Modelin tahmin yapabilmesi için kullanılan bağımsız değişkenlerdir. Bir özellik, verinin bir boyutunu veya niteliğini temsil eder (örneğin, bir evin fiyatını tahmin ediyorsak, "evin büyüklüğü" bir özellik olabilir).

5. **Etiket (Label)**:
   - Denetimli öğrenmede modelin öğrenmeye çalıştığı hedef değişken veya çıkıştır. Örneğin, ev fiyatları bir etiket olabilir.

6. **Aşırı Uyum (Overfitting)**:
   - Modelin, eğitim verisindeki gürültüyü veya rastlantısal desenleri öğrenmesi, böylece test verisinde düşük performans sergilemesi durumudur. Model, eğitim verisine çok iyi uyum sağladığında, genellikle genellenebilirlik kaybı yaşanır.

7. **Yetersiz Uyum (Underfitting)**:
   - Modelin, eğitim verisinin temel yapısını öğrenememesi durumudur. Modelin basitliği, veriyi yeterince temsil etmemesine ve düşük performans sergilemesine yol açar.

8. **Hiperparametre (Hyperparameter)**:
   - Modelin eğitim sürecini yöneten ve dışarıdan belirlenen parametrelerdir. Örneğin, öğrenme oranı veya karar ağacı derinliği gibi ayarlar hiperparametrelerdir.

9. **Özellik Mühendisliği (Feature Engineering)**:
   - Ham verilerden, modelin daha iyi öğrenmesini sağlayacak yeni özellikler türetme sürecidir.

#  Model Doğrulama Yöntemleri

Model doğrulama, makine öğrenmesi modellerinin genellenebilirliğini değerlendirme sürecidir. Bu, modelin yalnızca eğitim verisine değil, yeni ve bilinmeyen verilere karşı da ne kadar iyi performans gösterdiğini test etmeye yönelik bir adımdır.

### Yaygın Doğrulama Yöntemleri:

1. **Holdout Yöntemi**:
   - Veri kümesi, eğitim ve test veri kümesine bölünür. Model eğitim veri kümesinde eğitilir, test veri kümesinde ise doğrulanır. 
   - **Dezavantajı**: Eğitim ve test verisi arasındaki bölünme sonucu doğrulama süreci zaman zaman yanıltıcı olabilir.

2. **K-Fold Cross-Validation**:
   - Veri kümesi K eşit parçaya bölünür. Model, K-1 parça ile eğitilir ve kalan bir parça ile doğrulanır. Bu işlem K kez tekrarlanır ve her seferinde doğrulama sonucu kaydedilir.
   - **Avantajı**: Kümelenmiş veri setlerinde doğrulama için daha güvenilir sonuçlar verir.

3. **Leave-One-Out Cross-Validation (LOOCV)**:
   - Bu yöntem, küçük veri setleri için uygundur. Veri kümesindeki her örnek, sırasıyla test verisi olarak kullanılır ve geri kalan örneklerle eğitim yapılır.
   - **Avantajı**: Veriyi en iyi şekilde kullanma fırsatı tanır, ancak hesaplama açısından pahalıdır.

4. **Stratified K-Fold Cross-Validation**:
   - Özellikle dengesiz sınıflandırma problemleri için kullanılır. Veri setinin her sınıfını, her katmanda eşit şekilde temsil eder.
   - **Avantajı**: Modelin tüm sınıflarda tutarlı performans sergilemesini sağlar.

#  Yanlılık - Varyans Değiş Tokuşu

Yanlılık ve varyans, modelin doğruluğu üzerinde önemli etkiler yapar. Bu iki faktör arasında bir denge bulmak önemlidir.

### Yanlılık (Bias) ve Varyans (Variance) İlişkisi:

- **Yanlılık (Bias)**: Modelin, veriyi ne kadar iyi öğrenemediğinin bir ölçüsüdür. Yüksek yanlılık, modelin "underfitting" yapmasına neden olur.
  
- **Varyans**: Modelin eğitim verisine ne kadar duyarlı olduğunun ölçüsüdür. Yüksek varyans, modelin "overfitting" yapmasına yol açar.

### Yanlılık - Varyans Değiş Tokuşu:

- **Yüksek Yanlılık, Düşük Varyans**: Model çok basit ve genelleştirilebilir, ancak eğitim verisinde düşük performans gösterir. (Underfitting)
  
- **Düşük Yanlılık, Yüksek Varyans**: Model eğitim verisine çok iyi uyum sağlar ancak test verisi üzerinde kötü performans gösterir. (Overfitting)

Optimal model, düşük yanlılık ve düşük varyans seviyesine ulaşmayı hedefler, bu da hem doğru hem de genelleştirilebilir bir modelin oluşturulmasını sağlar.

# Parametre, Hiperparametre ve Model Optimizasyonu

Makine öğrenmesinde, doğru parametre ve hiperparametre ayarlarını bulmak modelin başarısını doğrudan etkiler. Model optimizasyonu, daha iyi sonuçlar almak için bu ayarları optimize etme sürecidir.

### Parametreler:

- Modelin öğrenmesi sırasında belirlediği değerlerdir. Örneğin, doğrusal regresyon modelindeki **ağırlıklar** ve **bias terimleri** parametrelerdir.

### Hiperparametreler:

- Modelin eğitim sürecini belirleyen ayarlardır. Örneğin, öğrenme oranı, karar ağacı derinliği gibi ayarlar hiperparametrelerdir.
  
### Model Optimizasyonu:

Model optimizasyonu, hiperparametrelerin doğru şekilde ayarlanmasını ifade eder. Grid Search, Random Search ve Bayesian Optimization gibi yöntemler ile en iyi hiperparametreler bulunabilir.

#### Grid Search:
- **Tüm olası hiperparametre kombinasyonlarını dener**. Yavaş ancak kesin sonuçlar verir.

#### Random Search:
- Hiperparametreleri **rastgele örnekler** ve daha hızlı sonuç verir.

#### Bayesian Optimization:
- Önceki denemelere göre iyileşir ve daha **verimli hiperparametreler** önerir.
